<h1><b>Email Simulation</b></h1>

<h4>This notebook will simulate the exchanging of emails between two people: Bob and Sue, with a server handling the storage and representation of the emails</h4>

<p>The program consists of 5 modules (newQueue, server, client, message, main)</p>

<ul>
    <li><b>newQueue</b> - A module that creates a FIFO queue with a Python List as the backing data structure</li>
    <li><b>server</b> - A module that simulates a server. It handles its data with a newQueue object and a CSV file. The server recieves messages from client objects and puts them in the queue. After every 4 seconds, the server will get each element of the queue and log it into the CSV file</li>
    <li><b>client</b> - A module that simulates an email. It contains the name of the email owner, the delay time set between automated message sends, and points to the server it is associated with.</li>
    <li><b>message</b> - A module that simulates an email message. It contains information on the sender of the message, the recipient of the message, and the content of the message. It also contains a __str__ method to assist in representing the message in the terminal.</li>
    <li><b>main</b> - This module contains the main method of the program. It initializes the server, the clients, establishes a counter for the amount of messages sent and the time elapsed since the server started. It then begins a loop that will send off 20 messages over a period of time. Every 4 seconds, each message sent to the server will be processed into a CSV file and will be printed in the terminal. Once 20 messages have been sent, the program will finish executing.</li>
</ul>
